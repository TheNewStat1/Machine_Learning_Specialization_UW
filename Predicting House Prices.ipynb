```python
# Launch Turi Create
import turicreate as tc
```

```python
# Load house sales data
sales = tc.SFrame('home_data.csv')
```

```python
# view the SFrame
sales
```

# Explore

```python
# Use .show() to view the dataset
sales.show()
```

```python
# specify ranges and filter & Build a scatter plot
tc.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

```python
## train simple regression model
# split data into a training and test set
training_set, test_set = sales.random_split(.8,seed=0)
print(training_set.shape, 'training set shape')
print(test_set.shape, ' test set shape')
```

```python
# train simple regression model
sqft_model = tc.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

```python
## Evaluate the quality of our model
# calculate average price of house in test set
print (test_set['price'].mean())
```

```python
# use .evaluate() to get root mean square error
print (sqft_model.evaluate(test_set))
```

```python
## Explore model a little further
# compute model coefficients
sqft_model.coefficients
```

```python
# Calculate predicted values of the model and 
# plot it agianst actual values from the test set
predicted_price = sqft_model.predict(test_set)
tc.visualization.scatter(predicted_price ,test_set['price'])
```

```python
# Explore other features of the data
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
```

```python
# .show() on the list of features
sales[my_features].show()
```

```python
# plot of zipcodes and house price
tc.show(sales['zipcode'],sales['price'])
```

```python
# Build a model with these additional features
my_features_model = tc.linear_regression.create(training_set,target='price',features=my_features)
```

```python
# Compare simple model with more complex one
print (sqft_model.evaluate(test_set), 'simple square foot model')
print (my_features_model.evaluate(test_set), ' more complex my_features model:', my_features)
```
